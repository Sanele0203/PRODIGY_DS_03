## Calculate the number of rows and columns for subplots

```python
num_plots=len(cat_cols)
num_rows=(num_plots+1)//2
num_cols=2
##Create a new figure
plt.figure(figsize=(20,25))
##Loop through each feature and create a counter plot
for i,feature in enumerate(cat_cols,1):
    plt.subplot(num_rows,num_cols,i)
    sns.countplot(x=feature,data=df,palette='Wistia')
    plt.title(f'Bar plot of:{feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
column=df[['age','campaign','duration']]
q1=np.percentile(column,25)
q3=np.percentile(column,75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df[['age','campaign','duration']]=column[(column>lower_bound)&(column<upper_bound)]
```

```python
high_corr_cols=['emp.var.rate','euribor3m','nr.employed']
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df_encoded=df1.apply(lb.fit_transform)
df_encoded
```

```python
x=df_encoded.drop('marital',axis=1)
y=df_encoded['marital']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print('Accuracy Score',acc)
    cn=confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cn)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
